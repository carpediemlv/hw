```python
#dropna() in df

#You have a dataset with missing values spread out across several columns. Each row represents a full observation, like survey answers. Before you analyze, you should make sure that every observation is complete.
#Using df.dropna() gets rid of rows with missing values, making sure that all of your observations are full. This method is better when missing data across rows could make the analysis less accurate, like when each observation needs to be full in order to run a certain statistical test or model.

#del df['col']

#You have a collection with several columns, but one column has a lot of missing values (for example, 90% of the values are missing). A lot of data would be lost if whole rows with missing numbers were taken out.
#Using del df['col'] gets rid of the problematic column and keeps the rest of the information. This method works best when a single column has a lot of missing data, making df.dropna() useless because it would get rid of too many rows.

#Using del df['col'] Before dropping df

#df.dropna() is used to clean up the leftover data after del df['col']. This gets rid of columns with too many missing values. This plan stops the loss of a lot of rows that might only be missing a few columns because they aren't full.
#If you get rid of the problematic columns first, you can keep more useful data for analysis because the remaining dataset will have more full observations.
```

```python
import pandas as pd

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df = pd.read_csv(url)

# Check missing values in the dataset
missing_before = df.isna().sum()
print("Missing values before cleaning:")
print(missing_before)
```

```python
# Identify columns with more than 50% missing values (adjust threshold as needed)
high_missing_cols = df.columns[df.isna().mean() > 0.5]

# Remove columns with high missing values
for col in high_missing_cols:
    del df[col]

# Drop rows with any remaining missing values
df_cleaned = df.dropna()

# Check missing values after cleaning
missing_after = df_cleaned.isna().sum()
print("Missing values after cleaning:")
print(missing_after)

# Display shape of the DataFrame before and after cleaning
print(f"Shape before cleaning: {df.shape}")
print(f"Shape after cleaning: {df_cleaned.shape}")
```

```python
#The main thing that makes them different is that they work in different ways. df.describe() finds the numbers that aren't missing in every column of the DataFrame as a whole. As df.groupby(“col1”)[“col2”].describe() works, it figures out what each group in the non-missing value col1 means. The missing number col2 will be taken out of each group. While both df.describe() and df.groupby(“col1”)[“col2”].describe() don't count missing values, using df.describe() affects the total number of non-missing values in all rows, while df.groupby(“col1”)[“col2”].describe() could give a different count for each group because of how the missing values are spread out in each group. When numbers are missing, they change the count of each group separately.
```
