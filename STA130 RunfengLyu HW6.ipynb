```python
# 3.
# The true line is based on the original values we set. 
# It shows the perfect connection between  and , with no random errors.
# Taking into account the random noise in the data, the fitting line shows the best line that fits through the collected data points.
# The fitted line might not match the true line exactly because of this, but it should be close if the sample size is big enough. 
# The effect of sampling variability can be seen in the slightly different fitted lines that are produced by repeated trials.
# This code and description show how random sampling changes regression: 
# the true model parameters don't change, but each sample gives us different data points, 
# which means that the fitting line is also slightly different.
```

```python
# 4.
# It is in fitted_model.params that you can find the predicted parameters, mainly the intercept and slope. 
# These are used to get the numbers that fit the fitted_model in a simple linear regression model. 
# The sample data is used to fit the model by reducing the sum of the squared differences between the real and predicted values of . 
# This gives us these values of the parameters. As soon as the model has been fitted, the method is used to find the predicted values, 
# which are also known as fitting values, for each item in the collection. 
# As for the real model, it has an error term and no one knows what the values are. 
# It's not possible to know what the factors are in the fitted model, and random mistake is not taken into account in any of the predictions. 
# Since there is a straight line between  and  in the sample data, 
# fitted_model.fittedvalues shows the most accurate predictions that can be made. 
```

```python
# 5.
# The "ordinary least squares" (OLS) method for linear regression picks the line that 
# makes the sum of the squared residuals as small as possible. 
# The gap between what was seen and what the model said it would be is called a residual. 
# Instead of using the residuals themselves, the OLS method uses the square of each residual. 
# This way, negative numbers don't cancel out positive ones, so all differences are taken into account evenly. 
# The OLS method finds the line that best fits the data by minimizing the sum of these squared residuals. 
# This reduces the distance between the measured points and the line, 
# giving the most accurate linear estimate of the relationship between X and Y.
```

```python
# 6.
# The formula is the percentage of the change in Y (the result variable) that can be explained by the regression model. 
# Compare the total change in Y (relative to its mean) to the change in the residuals, 
# which are the differences between the values of Y that were measured and those that were forecasted (Y^). 
# The residual variation will be small if the model is good at capturing most of the variation in Y. 
# This means that R^2 will be close to 1, which means the model is a good fit. 
# This R^2 number is given directly by fitted_model.rsquared in statsmodels. np.corrcoef(Y, fitted_model.fittedvalues)[0,1] is also used.
# To find R^2, **2 squares the correlation between Y and the expected values Y^, 
# since the square of this correlation also shows how well the model can explain things. 
# Finally, np.corrcoef(Y, x)[0,1]**2 gives the square of the correlation between Y and X, 
# which shows how strong their straight-line relationship is. 
# This number is like R^2 in simple linear regression because the model only uses X to explain the changes in Y.
```

```python
# 10.
import numpy as np
import pandas as pd
import plotly.express as px
import statsmodels.formula.api as smf
from scipy.stats import norm

long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit
long_wait_data = old_faithful[long_wait]

n_bootstraps = 1000  # Number of bootstrap samples
bootstrapped_slope_coefficients = []

for _ in range(n_bootstraps):
    sample = long_wait_data.sample(n=len(long_wait_data), replace=True)
    model = smf.ols('duration ~ waiting', data=sample).fit()
    bootstrapped_slope_coefficients.append(model.params['waiting'])

bootstrap_conf_interval = np.percentile(bootstrapped_slope_coefficients, [2.5, 97.5])

n_simulations = 1000 
simulated_slope_coefficients = []

for _ in range(n_simulations):
    simulated_data = long_wait_data.copy()
    simulated_data['duration'] = 1.65 + 0 * simulated_data['waiting'] + \
                                 norm.rvs(0, 0.37, size=len(simulated_data))
    model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    simulated_slope_coefficients.append(model.params['waiting'])

simulated_mean_slope = np.mean(simulated_slope_coefficients)
p_value = (np.abs(simulated_slope_coefficients) >= np.abs(simulated_mean_slope)).mean()

print("Bootstrap 95% Confidence Interval for Slope:", bootstrap_conf_interval)
print("Mean Slope under Null Hypothesis:", simulated_mean_slope)
print("p-value for Slope Coefficient under Null Hypothesis:", p_value)
```

```python
# 11.

# Models 1, 2, and 3 look at waiting as a continuous variable and try to figure out 
# if there is a continuous linear relationship between waiting and length across a range of values. 
# But the indicator variable approach sees waiting as either "short" or "long" 
# and compares the average differences in how long people wait in these two groups.
# Models 2 and 3 look at parts of the data (like short or long waits) to find possible differences in the links between these groups. 
# The indicator variable model, on the other hand, compares the mean difference across the whole dataset.


# The null hypothesis for the indicator variable model is: 
# H: There is no difference in the average eruption duration between "short" and "long" wait times.

# In this context, rejecting the null hypothesis would suggest that the average eruption duration differs between short and long waits, 
# while failing to reject it would imply no significant difference between the groups.

# If the p-value for kind is big (e.g., p < 0.05), 
# it means that the null hypothesis is false and the eruption lasts a lot longer during long wait times compared to short wait times.
# If the p-value is not significant, we cannot dismiss the null hypothesis. 
# This means that there is no significant difference between the two groups in the average length of time.
# Instead of looking at a straight line connection, 
# this method helps us figure out if "long" waits have a different average length of time than "short" waits.
```

```python
# 13.(A)

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

old_faithful = sns.load_dataset('geyser')

long_wait_limit = 68
old_faithful['kind'] = old_faithful['waiting'].apply(lambda x: 'short' if x < long_wait_limit else 'long')

observed_diff = old_faithful.groupby('kind')['duration'].mean().diff().iloc[-1]

n_permutations = 1000
permutation_diffs = []

for _ in range(n_permutations):
    shuffled_kind = old_faithful['kind'].sample(frac=1, replace=False).reset_index(drop=True)
    shuffled_data = old_faithful.assign(kind_shuffled=shuffled_kind)
    permuted_diff = shuffled_data.groupby('kind_shuffled')['duration'].mean().diff().iloc[-1]
    permutation_diffs.append(permuted_diff)

permutation_diffs = np.array(permutation_diffs)

p_value = np.mean(np.abs(permutation_diffs) >= np.abs(observed_diff))

plt.hist(permutation_diffs, bins=30, density=True, alpha=0.7, label='Permutation Differences')
plt.axvline(observed_diff, color='red', linestyle='--', label='Observed Difference')
plt.title('Permutation Test for Mean Difference in Eruption Duration')
plt.xlabel('Mean Difference')
plt.ylabel('Density')
plt.legend()
plt.show()

print("Observed Mean Difference:", observed_diff)
print("p-value:", p_value)
```

```python
# 13.(B)

import numpy as np
import pandas as pd
import seaborn as sns

old_faithful = sns.load_dataset('geyser')

long_wait_limit = 68
old_faithful['kind'] = old_faithful['waiting'].apply(lambda x: 'short' if x < long_wait_limit else 'long')

short_group = old_faithful[old_faithful['kind'] == 'short']['duration']
long_group = old_faithful[old_faithful['kind'] == 'long']['duration']

n_bootstraps = 1000
bootstrapped_mean_differences = []

for _ in range(n_bootstraps):
    short_sample = short_group.sample(n=len(short_group), replace=True)
    long_sample = long_group.sample(n=len(long_group), replace=True)
    mean_difference = long_sample.mean() - short_sample.mean()
    bootstrapped_mean_differences.append(mean_difference)

bootstrapped_mean_differences = np.array(bootstrapped_mean_differences)
confidence_interval = np.quantile(bootstrapped_mean_differences, [0.025, 0.975])

print("95% Bootstrap Confidence Interval for Mean Difference:", confidence_interval)
```

```python
# 13.(a)

# In a random test, we look at the idea that there isn't a change in the average length of time between the "short" and "long" wait groups.
# We randomly mix up the "short" and "long" group names and find the average difference in lengths for each mixed dataset. 
# This gets rid of any real link between group type and length, 
# so the data would be like it would be if there were no group effect.
# If you do this many times, you will get a range of mean differences under the null hypothesis. 
# After that, the mean change will be compared that we saw to this range. 
# If the difference that was seen is very big compared to this range, then it's not likely that it happened by chance, 
# which is evidence against the null hypothesis. 
# It doesn't test a theory; instead, it makes a confidence range for the difference in means between the "short" and "long" groups. 
# It will take different samples from within each group over and over again to make "bootstrap samples" 
# and find the mean difference between each pair of samples. 
# Based on what was seen in the data format, this process makes a range of mean differences. 
# The lower and higher limits of a 95% confidence interval are the 2.5th and 97.5th percentiles of this distribution. 
# This gives us an idea of the range for the real mean difference.
```

```python
# 13.(b)

# Three different ways are used to compare the lengths of eruptions during the "short" and "long" wait times: the random test, 
# the bootstrap confidence interval, and the indicator variable model.
# The difference between groups is measured in some way by each method. 
# A regression value is used by the indicator variable model, 
# while sampling distributions are used by the permutation test and bootstrap confidence interval methods.

# It uses parameters, like the indicator variable model. 
# The mean difference is estimated directly as a coefficient for the kind variable (indicator). 
# It is based on a linear model with error terms that are normally distributed.
# Not all tests are parametric, like the randomization test. 
# A random mixing method is used to make a distribution under the null hypothesis instead of assumptions about how the data is distributed.
# The bootstrap method is also non-parametric, but instead of switching labels, 
# it resamples within each group to get an idea of how variable the sampling is at the mean difference. 
# Instead of testing a null hypothesis, this gives us a confidence range. With the p-value for checking significance, 
# the indicator variable model gives you an expected mean difference.
# Compare the measured mean difference to the permutation distribution to get a p-value from the permutation test.
# Although it doesn't directly test a theory, the bootstrap method gives a confidence range for the mean difference.
```
